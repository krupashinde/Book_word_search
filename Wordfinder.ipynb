```python
from google.colab import files
import io
import re
import shutil
from IPython.display import display, HTML

# Step 1: Upload
print("üìÅ Upload your book (.txt) file:")
uploaded = files.upload()
file_name = list(uploaded.keys())[0]
book_path = "/content/book.txt"
shutil.move(file_name, book_path)

# Step 2: Read text
with io.open(book_path, 'r', encoding='utf-8', errors='ignore') as f:
    lines = f.readlines()

full_text = ''.join(lines)
sentences = re.split(r'(?<=[.!?])\s+', full_text)
lines_per_page = 40

# Step 3: Get keyword
keyword = input("Enter the word or phrase you remember: ").strip().lower()

# Step 4: Find matching sentences
def find_starting_line(sentence):
    for i, line in enumerate(lines):
        if sentence[:30].lower() in line.lower():
            return i + 1
    return -1

results = []
for sent in sentences:
    if keyword in sent.lower():
        clean_sent = sent.strip()
        line_number = find_starting_line(clean_sent)
        if line_number != -1:
            page_number = (line_number - 1) // lines_per_page + 1
            # üî≤ HTML-based highlight: black background
            highlighted = re.sub(
                f"({re.escape(keyword)})",
                r"<span style='background-color:black; color:white; padding:2px;'>\1</span>",
                clean_sent, flags=re.IGNORECASE
            )
            results.append((page_number, line_number, highlighted))

# Step 5: Display nicely
if results:
    html_output = "<h3>üìñ Search Results</h3><hr>"
    for i, (page, line, sentence) in enumerate(results[:10]):
        html_output += f"<p><b>üîπ Match #{i+1}</b><br>üìÑ Page: {page} |  Line: {line}<br>üó®Ô∏è  {sentence}</p>"
    if len(results) > 10:
        html_output += f"<p>...and {len(results) - 10} more matches found.</p>"
    display(HTML(html_output))
else:
    print("‚ùå No matching sentences found. Try another word.")
```
